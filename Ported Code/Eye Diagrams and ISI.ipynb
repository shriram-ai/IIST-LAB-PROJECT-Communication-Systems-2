```python
import matplotlib.pyplot as plt
import numpy as np
from helper_functions import *
```

Channel Specifications

```python
f_pass = 10
gain = 1
f_stop = 20
filter_length = 100
```

Generating Baseband BPSK signal 

```python
N = 100
Tb = 0.1
Fs = 100
bitstr = bitgen(N)

line_code = pam_code(bitstr, Tb, Fs)
plt.plot(line_code)
```

Channel Magnitude Response and Output

```python
ax = plt.subplot()
ax, channel_output = channel(line_code, gain, f_pass, f_stop, Fs, filter_length, 1, ax)
```

Eye Diagram of Baseband BPSK (No ISI)

```python
ax = plt.subplot()
eye_diagram(line_code, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
plt.show()
```

Eye Diagram of Channel Output

```python
ax = plt.subplot()
eye_diagram(channel_output, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
plt.show()
```

Matched Filtering of Received Signal

```python
m_filter = np.ones(np.uint16(Tb*Fs))
filter_output = np.convolve(channel_output, m_filter)
```

Eye-diagram of Matched Filtered Received signal

```python
ax = plt.subplot()
eye_diagram(filter_output, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
plt.show()
```
