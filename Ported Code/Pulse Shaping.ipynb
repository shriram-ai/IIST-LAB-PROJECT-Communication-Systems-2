```python
import matplotlib.pyplot as plt
import numpy as np
from helper_functions import *
```

Channel Specifications

```python
f_pass = 10
gain = 1
f_stop = 20
filter_length = 100
```

Generating random sequence of bits

```python
N = 100
Tb = 0.1
Fs = 100
bitstr = bitgen(N)
```

Pulse Shaping at Transmitter - Baseband BPSK

```python
rect_line_code = pam_code(bitstr, Tb, Fs)
plt.plot(rect_line_code)
```

Eye Diagram - Rectangular pulse shape

```python
ax = plt.subplot()
eye_diagram(rect_line_code, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
```

PSD - Rectangular pulse shape

```python
ax = plt.subplot()
ax = PSD(rect_line_code, Fs, N*Tb, 'PSD', 1, ax)
```

Pulse Shaping - Sinc 

```python
sinc_line_code = sinc_code(bitstr, Tb, Fs)
plt.plot(sinc_line_code)
```

Eye Diagram - Sinc Pulse shape

```python
ax = plt.subplot()
eye_diagram(sinc_line_code, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
```

PSD - Sinc pulse shape

```python
ax = plt.subplot()
ax = PSD(sinc_line_code, Fs, N*Tb, 'PSD', 1, ax)
```

Pulse Shaping - Raised Cosine 

```python
rc_line_code = raised_cos_code(bitstr, Tb, Fs, 1)
plt.plot(rc_line_code)
```

Eye Diagram - Raised Cosine pulse shape

```python
ax = plt.subplot()
eye_diagram(rc_line_code, Tb, Fs, [-0.5*Tb, 1.5*Tb], 'Eye Diagram', ax = ax)
```

PSD - Raised Cosine pulse shape

```python
ax = plt.subplot()
ax = PSD(rc_line_code, Fs, N*Tb, 'PSD', 1, ax)
```

Channel - Magnitude Response

```python
ax = plt.subplot()
ax, channel_output = channel(rc_line_code, gain, f_pass, f_stop, Fs, filter_length, 1, ax)
```

PSD of Channel Output - Rectangular pulse shape

```python
ax, channel_output = channel(rect_line_code, gain, f_pass, f_stop, Fs, filter_length, 0)
ax = plt.subplot()
ax = PSD(channel_output, Fs, N*Tb, 'PSD of Channel Output', 1, ax)
```

PSD of Channel Output - Sinc pulse shape

```python
ax, channel_output = channel(sinc_line_code, gain, f_pass, f_stop, Fs, filter_length, 0)
ax = plt.subplot()
ax = PSD(channel_output, Fs, N*Tb, 'PSD of Channel Output', 1, ax)
```

PSD of Channel Output - Raised Cosine pulse shape

```python
ax, channel_output = channel(rc_line_code, gain, f_pass, f_stop, Fs, filter_length, 0)
ax = plt.subplot()
ax = PSD(channel_output, Fs, N*Tb, 'PSD of Channel Output', 1, ax)
```
